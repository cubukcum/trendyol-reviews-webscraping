```python
# For using without the params file
import requests
import csv

# URL and the headers
url = "https://public-mdc.trendyol.com/discovery-web-websfxsocialreviewrating-santral/product-reviews-detailed"
headers = {
    "Accept": "application/json, text/plain, */*",
    "Accept-Encoding": "gzip, deflate, br, zstd",
    "Accept-Language": "en-US,en;q=0.9",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36",
}

# parameters for the request
params = {
    "sellerId": "968",
    "contentId": "65149494",
    "page": "1",
    "order": "DESC",
    "orderBy": "Score",
    "channelId": "1"
}

# Function to fetch reviews for a given page
def fetch_reviews(page):
    params["page"] = str(page)
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        data = response.json()
        return data["result"]["productReviews"]["content"]
    else:
        print(f"Failed to retrieve reviews for page {page}")
        return []

# reviews from pages 1 and 2
all_reviews = []
for page in range(1, 50):
    reviews = fetch_reviews(page)
    all_reviews.extend(reviews)

# CSV file name
csv_file = 'reviews.csv'

# field names (column headers)
fieldnames = ["User", "Rating", "Comment", "Date", "Seller", "Likes"]

# Write data to CSV file
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    # Write the header
    writer.writeheader()
    
    # Write the data
    for review in all_reviews:
        writer.writerow({
            "User": review['userFullName'],
            "Rating": review['rate'],
            "Comment": review['comment'],
            "Date": review['commentDateISOtype'],
            "Seller": review['sellerName'],
            "Likes": review['reviewLikeCount']
        })

print(f'Data has been written to {csv_file}')
```

```python
# works with the params file and creates a different csv for each product
import requests
import csv
import json

# URL and the headers
url = "https://public-mdc.trendyol.com/discovery-web-websfxsocialreviewrating-santral/product-reviews-detailed"
headers = {
    "Accept": "application/json, text/plain, */*",
    "Accept-Encoding": "gzip, deflate, br, zstd",
    "Accept-Language": "en-US,en;q=0.9",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36",
}

# Function to fetch reviews for a given page
def fetch_reviews(params):
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        data = response.json()
        return data["result"]["productReviews"]["content"]
    else:
        print(f"Failed to retrieve reviews for contentId {params['contentId']} page {params['page']}")
        return []

# Read the parameters from the text file
with open('/kaggle/input/d/mehmetcubukcu/products/params.txt', 'r') as file:
    lines = file.read().split('\n}\n')

# Process each set of parameters
for line in lines:
    if line:
        params = json.loads(line + '}')
        all_reviews = []
        
        # Collect reviews from pages 1 and 2
        for page in range(1, 3):
            params["page"] = str(page)
            reviews = fetch_reviews(params)
            all_reviews.extend(reviews)
        
        # Define the CSV file name based on contentId
        csv_file = f'reviews_{params["contentId"]}.csv'
        
        # Define the field names (column headers)
        fieldnames = ["User", "Rating", "Comment", "Date", "Seller", "Likes"]
        
        # Write data to CSV file
        with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            
            # Write the header
            writer.writeheader()
            
            # Write the data
            for review in all_reviews:
                writer.writerow({
                    "User": review['userFullName'],
                    "Rating": review['rate'],
                    "Comment": review['comment'],
                    "Date": review['commentDateISOtype'],
                    "Seller": review['sellerName'],
                    "Likes": review['reviewLikeCount']
                })
        
        print(f'Data has been written to {csv_file}')
```
